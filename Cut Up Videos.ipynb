# We will go through each video, and if it is greater than 5 seconds, will split it into 2 videos - one that is 3 seconds long and one that is the rest (at least 2 seconds). This will help increase the number of data points we have (through more opportunities for augmentation) and also smoothen out model fluctuations in training. 

```python
import moviepy, cv2, os 
from moviepy.video.io.VideoFileClip import VideoFileClip
LIMIT = 3

num = 0 
for directory in ['shorter_armflapping', 'shorter_control']: 
    for video_name in os.listdir("behavior_data/" + directory):
        if video_name == ".DS_Store": continue
        video = VideoFileClip(f"behavior_data/{directory}/{video_name}")
        if video.duration > LIMIT + 1: 
            # cut it up 
            start_video = video.subclip(0, LIMIT)
            start_video.write_videofile(f"behavior_data/{directory}/_{num}.mp4", audio_codec='aac')
            num += 1
            end_video = video.subclip(LIMIT, video.duration)
            end_video.write_videofile(f"behavior_data/{directory}/_{num}.mp4", audio_codec='aac')
            num += 1
        video.close() 

        os.remove(f"behavior_data/{directory}/{video_name}")
```
